```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
case = pd.read_csv('교통사고+현황(사고유형별)_2022.csv')
plt.rcParams.update({'font.family':'Malgun Gothic'})
case
```

```python
place = input('자치구 : ')
case1 = case.query('`자치구별(2)`== @place')\
            .transpose()\
            .iloc[4:]
case1.replace('-', 0, inplace = True)
case1.columns=['발생건수','부상자수']
case1.index = ['차대사람','차대차','차량단독','건널목']
case1 = case1.astype({'발생건수' : int, "부상자수" : int})
case1.plot.barh()
plt.title(f'{place} 유형별 사고량')
```
