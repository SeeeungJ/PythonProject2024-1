```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
acc = pd.read_csv('시간대별+교통사고+현황_2022.csv')
plt.rcParams.update({'font.family':'Malgun Gothic'})
acc.sort_values('자치구별(2)')
```

```python
place = input("자치구를 입력하세요 : ")
acc_gangnam = acc.query('`자치구별(2)`==@place')\
                .transpose()\
                .iloc[4:]
acc_gangnam.columns=['발생건수','사망자수','부상자수']
acc_gangnam.drop(columns = '사망자수', inplace = True)
acc_gangnam.index = [f"{str(i*2).zfill(2)}시" for i in range(1, 13)]

acc_gangnam = acc_gangnam.astype({'발생건수' : int, "부상자수" : int})
sns.lineplot(data = acc_gangnam, dashes = False)
plt.title(f'{place} 시간대별 사고량')
plt.figure(figsize=(14, 8))
```

```python
place = "강남구"
acc_gangnam = acc.query('`자치구별(2)`==@place')\
                .rename(columns 
                = {
                   '2023.1' : '00시',
                  '2023.2' : '02시',
                  '2023.3' : '04시',
                  '2023.4' : '06시',
                  '2023.5' : '08시',
                  '2023.6' : '10시',
                  '2023.7' : '12시',
                  '2023.8' : '14시',
                  '2023.9' : '16시',
                  '2023.10' : '18시',
                  '2023.11' : '20시',
                  '2023.12' : '22시'},
                index = 
                    {71 : '발생건수',
                   72 : '사망건수',
                   73 : '부상자수'})\
                .transpose()\
                .iloc[4:]

acc_gangnam.replace('-', 0, inplace = True)
acc_gangnam.to_csv('@.csv')
acc_gangnam
```

```python
gangnam_발생 = acc_gangnam['발생건수'].astype(int)
sns.lineplot(data= gangnam_발생)\
    .set(ylim=(0,800))
plt.title('2023년 강남구 시간별 사고 발생 건수')
plt.ylabel('건수')
```

```python
gangnam_사망 = acc_gangnam['사망건수'].astype(int)
sns.lineplot(data= gangnam_사망)\
    .set(ylim=(0,800))
plt.title('2023년 강남구 시간별 사망 발생 건수')
plt.ylabel('건수')
```

```python
gangnam_부상 = acc_gangnam['부상자수'].astype(int)
sns.lineplot(data= gangnam_부상)
plt.title('2023년 강남구 시간별 부상 발생 건수')
plt.ylabel('건수')
```

```python
sns.lineplot(data = [gangnam_발생,gangnam_부상], dashes = False)
plt.title('2023년 강남구 시간별 발생/부상 발생 건수')
plt.ylabel('건수')
plt.xlabel('시간')
```
