```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 교통량 데이터 불러오기
traffic_data = traffic_all

# 2시간 단위로 교통량 데이터 합산
traffic_data_hourly = pd.DataFrame()
time_intervals = [(i, i+1) for i in range(0, 24, 2)]

for start, end in time_intervals:
    col_name = f'{start}시-{end+1}시'
    traffic_data_hourly[col_name] = traffic_data.iloc[:, 4+start] + traffic_data.iloc[:, 4+end]

# 자치구를 포함하여 데이터프레임 구성
traffic_data_hourly['자치구'] = traffic_data['시군구']

# 사고량 데이터 불러오기
accident_data = accident_1

# 교통량과 사고량 데이터의 자치구 순서를 맞추기 위해 인덱스 재설정
accident_data_reordered = accident_data.set_index('자치구별(2)').reindex(traffic_data['시군구']).reset_index()

# 데이터 타입 점검 및 변환
traffic_data_hourly.drop(columns='자치구', inplace=True)
traffic_data_hourly = traffic_data_hourly.apply(pd.to_numeric, errors='coerce')

accident_data_reordered['소계'] = pd.to_numeric(accident_data_reordered['소계'], errors='coerce')

# 교통량과 사고량 간의 상관관계 계산
correlation = traffic_data_hourly.corrwith(accident_data_reordered['소계'])

# 상관관계 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=correlation.index, y=correlation.values)
plt.title('교통량과 사고 발생 건수의 상관관계')
plt.xlabel('시간대')
plt.ylabel('상관관계')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

# 교통량 데이터
traffic_data = traffic_all[['시군구', '전일']]
traffic_data.columns = ['시군구', '교통량']  # 컬럼명을 변경합니다.

# 사고량 데이터
accident_data = accident_1[['자치구별(2)', '소계']]
accident_data.columns = ['시군구', '사고량']  # 컬럼명을 변경합니다.

# 데이터프레임 병합
df_merged = pd.merge(traffic_data, accident_data, on='시군구')

# 교통량과 사고량 간의 상관관계 계산
correlation = df_merged['교통량'].astype('int').corr(df_merged['사고량'].astype('int'))

# 결과 출력
print("교통량과 사고량 간의 상관관계:", correlation)
```
